## 0. Abstract
딥러닝 알고리즘, 특히 CNN은 메디컬 이미지를 분석하기 위한 방법론으로써 빠르게 선택받고 있는 추세이다. 
+ 이 논문은 메디컬 이미지 분석에 적합한 주요 딥러닝 개념들을 리뷰하고 이 분야에서 주로 지난해 발행된 300개 이상의 논문들을 요약한다.
+ Image Classification, Object Detection, Segmentation, Registration 등등에 사용된 딥러닝 사용법도 서베이한다.
+ 다양한 응용분야(Neuro, Retinal, Pulmonary, Digital Pathology, Breast, Cardiac, Abdominal, Musculoskeletal)마다 수행된 연구들을 요약한다.
+ 마지막으로 최근 업계상황 요약, 중요한 논쟁거리, 그리고 미래 연구에 대한 방향성을 제시한다.

## 1. Introduction

메디컬 이미지를 스캔해서 컴퓨터에 옮길 수 있게 되면서 연구자들은 자동 분석 시스템을 구축해왔다.

+ **1970 ~ 1990년대** : 이 시대의 메디컬 이미지 분석은 Low-level Pixel Processing을 반복 적용해서 행해졌고, 복합적인 규칙기반시스템(Rule-based System)을 구축하기 위해 수학적 모델을 사용해 특정 업무를 수행했고, 그 당시 인공지능계에서 인기 있었던 것 처럼 if-then-else문이 많이 사용되었다. 이런 규칙기반시스템은 GOFAI(옛날 방식의 인공지능)이었고 많이 불안정했다.

    ※ 규칙기반시스템 : 사람이 미리 정한 규칙에 의거해 if-else문으로 구성된 시스템
  
  
+ **1990년대 말** : 시스템 개발을 위해 Training Data를 사용하는 Supervised Technique이 메디컬 이미지 분석에 사용되었다. Active Shape Model, Atlas Method, 그리고 Feature Extraction 개념과 Statistical Classifier가 사용되었다. 이런 패턴인식 / 머신러닝 접근법은 여전히 매우 인기 있고 상업적으로 성공한 메디컬 이미지 분석 시스템의 근간을 이룬다.


<br>여기까지 '사람이 디자인한 시스템'에서 '컴퓨터가 데이터를 학습해 만든 시스템'으로의 변화가 있었다. 컴퓨터의 알고리즘은 고차원 Feature Space에서 최적의 Decision Boundary를 결정할 수 있다. 이런 시스템을 디자인할 때 중요한 단계는 이미지의 **Discriminant Feature**을 추출하는 일이다. 근데 이 과정은 여전히 사람에 의해 수행되고 있고 어떤 사람들은 이걸 **Handcrafted Feature**라고 부른다.


<br>다음 단계는 컴퓨터가 문제 해결을 위해 스스로 가장 잘 데이터를 표현할 수 있게 특징들을 학습시키는 작업이다. 이 개념은 많은 딥러닝 모델들의 바탕이 된다. 이미지를 받아 높은 차원의 Feature를 학습해 질병 등을 찾아내는 DNN 알고리즘 등..


지금까지 이미지 분석에 가장 성공적인 모델은 CNN이다. CNN은 Input을 작은 필터들로 변형시키는 많은 레이어를 갖고 있다.


+ **1970** : CNN 연구가 시작됨 *(Fukushima, 1980)*
+ **1995** : 메디컬 이미지 분석에 적용됨 *(Lo et al.)*
+ **1998** : LeNet으로 손글씨를 분석해 CNN이 실생활에 사용될 수 있다는 걸 보여줌 *(LeCun et al.)*
+ **2012** : AlexNet이 20120 ImageNet Challenge 우승 *(Krizhevsky et al.)*
+ **2014** : 더 깊은 CNN의 도입 *(Russakovsky et al.)*

<br>메디컬 이미지 분석 업계에서는 이런 굵직한 발전이 없었고, 'Handcrafted Feature'에 의한 분석이 많았다.

+ AlexNet의 등장 이전 메디컬 이미지 분석에 관한 서베이 : Bengio et al.(2013)
+ Health Informatics에 관한 넓은 범위의 딥러닝 서베이 : Ravi et al. (2017)
+ 2015~16부터 급격히 관련 논문들이 쏟아지기 시작 : Shen et al. (2017)
+ 저 논문도 Retinal 쪽은 다루지 않는 등 완전하진 않아서 우리가 이 서베이를 썼다! (Abstract 참조)

## 2. Overview of Deep Learning Methods

메디컬 이미지 분석에서 사용되는 알고리즘들의 컨셉과 구조에 대해 설명한다.

### (1) Learning Algorithms

머신러닝 방법론은 대충 Supervised랑 Unsupervised로 나눌 수 있다.
+ Supervised : (x, y)로 Labeling된 데이터를 학습시키는 과정 (= 가중치 업데이트)
+ Unsupervised : Label이 없고 데이터의 패턴을 찾는 과정 (PCA, Clustering 등.. Loss Function 최적화시킴)


### (2) Neural Networks

신경망은 대부분의 딥러닝 알고리즘의 근간을 이루며, 활성함수 a와 파라미터 W, B가 들어있는 뉴런들로 구성되어 있다. 활성함수 \\({a} = {\sigma}({W}^{T}*{X}+{B})\\)에서 \\({W}\\)는 가중치, \\({B}\\)는 편차, 그리고 \\({\sigma}\\)는 Transfer Function으로서 일반적으로 sigmoid나 hyperbolic tangent 함수를 의미한다.

신경망의 마지막 Layer는 Softmax Function을 씌워 어떤 클래스에 속하는지에 대한 값을 내보낸다. 

Stochastic Gradient Descent는 파라미터 \\({W}\\)와 \\({B}\\)를 데이터셋에 Maximum Likelihood하게 조정하기 위한 가장 잘 쓰이는 방법이다. 한 번 GD를 적용할 때 'mini-batch'라고 해서 전체 데이터셋이 아닌 일부를 사용해서 값을 갱신한다. 이건 Negative Log-likelihood를 최소화 하는 것과 같은 의미인데, 이 결과는 Binary 및 Multi-class 문제에서의 Cross-Entrophy이다.

+ \\({W}\\)는 Row가 input의 차원, Column이 output의 차원으로 이루어진 Matrix이다.
+ Input과 output 사이의 Layer를 'Hidden Layer'라고 부른다.
+ 신경망이 이 Hidden Layer를 가지고 있는 경우 DNN이라고 부른다.

<br>*※ Transfer Funtion를 보통 Activation Function이라고 부르고 sigmoid나 tanh 말고도 ReLU가 쓰인다고 알고있는데, 여기선 a를 Activation Function으로 명명했기에 Transfer Function이라는 용어를 사용한 것 같음*

<br>오랜 시간 동안 DNN은 효과적으로 학습시키기 어렵다고 여겨졌다. 하지만 2006년, DNN을 층 별로 Unsupervised하게 Pre-training시키고, 나머지 신경층은 Supervised하게 Fine-tuning시키는 것이 효과적인 학습 방법일 수 있다는 결과가 나왔다. 이런 방식이 사용된 두 가지 아키텍쳐는 SAE와 DBN이다. 하지만 이런 기술은 굉장히 복잡하고 만족스러운 결과를 내기까지 수많은 수정을 가해야 한다.

최근 가장 인기 있는 CNN, RNN 모델은 Supervised 방식으로 끝까지 훈련시키되 훈련 과정을 크게 단순화시키는 방식을 취한다. CNN은 현재 메디컬 이미지 분석에서 가장 널리 쓰이고 있고, RNN도 꽤 인기 있다.

## 3. Deep Learning Uses in Medical Imaging

### (1) Classification

#### ① Image/Exam Classification
Image/Exam Classification(이미지/검사 분류)은 메디컬 이미지 분석에 딥러닝이 가장 주요한 역할을 하는 분야 중 하나이다. Exam Classification에서 일반적으로 input으로 하나 이상의 Image/Exam을 넣고, output으로 하나의 진단 결과를 받는다. 모든 Exam들은 샘플이 되고 보통 데이터셋의 크기는 컴퓨터비전에 비해 작은 편이다. 따라서 Transfer Learning이 널리 사용된다.

Transfer Learning은 학습에 큰 데이터셋을 필요로 하는 작업에 Pre-trained Network를 사용하는 것이다. 이 방법엔 두 가지 전략이 있다.

+ Pre-trained Network를 **Feature Extractor**로 사용
+ Pre-trained Network를 **Fine-tuning**

두 전략 모두 널리 사용되지만, Antony et al.(2016)과 Kim et al.(2016)의 논문에 어떤 방식이 더 좋은지에 대한 결과가 나와 있다. 근데 둘의 결과가 정반대임;

아무튼 Fine-tuning을 메디컬 데이터 분석에 사용해 전문의 수준의 정확도를 이뤄낸 두 논문이 있으니 읽어보면 좋을 것이다. (Esteva et al.,2017; Gulshan et al.,2016)

<br>메디컬 이미징계는 앞서 말한 SAE나 RBM 사용에 집중했었다. Brosch and Tam(2013), Plis et al.(2014), Suk and Shen(2013), and Suk et al.(2014)에서는 MRI를 활용한 알츠하이머 환자 진단에 SAE와 DBN을 사용했다. 

하지만 **지금은 CNN쪽으로 기울고 있다**. 2015~17년에 발표된 47편의 논문들 중 36편이 CNN을 사용했고 5개가 AE, 6개가 RBM을 사용했기 때문이다. 이 방식들은 MRI부터 CT까지 다양한 방면에 사용되고 있다.

+ 알츠하이머에 2D가 아닌 3D Convolution을 사용 : Hosseini-Asl et al.,2016; Payan and Montana,2015
+ CNN을 MRI DTI에서 나온 뇌 연결 그래프에 적용 : Kawahara et al., 2016

이 사람들은 각각 edge-to-edge, edge-to-node, node-to-graph라는 세 층의 layer를 기반으로 신경망을 구성했다. 그들은 이 신경망으로 뇌발달을 예측했고 뛰어난 결과물을 냈다.

<br>요약하자면, Exam Classification에서 **CNN은 현재 가장 표준 기술**임에 틀림없다. 특히 **Pre-trained된 CNN**은 어떤 면에선 전문의를 뛰어넘는 놀라운 결과를 냈다.

<br>
#### ② Object or Lesion Classification
메디컬에서의 Object Classification은 이미지의 작은 부분을 두 개 혹은 그 이상의 클래스로 나누는 일이 보통이다. (흉부 CT에서 결절의 갯수 알아내기 등) ***단일 병변의 모양에서 지역적 정보***를 얻어내거나, ***병변의 위치로부터 전체적인 정보***를 얻어내는 이런 작업들은 정확한 Classification이 요구된다! 

##### Multi-Stream CNN
이런 일들을 하려면 일반적인 딥러닝 구조로는 불가능했기에, 몇몇 저자들은 이 문제를 multi-scale로 풀어내기 위한 multi-stream 구조를 사용했다.

+ Shen et al., (2015) : 각기 다른 크기로 들어온 결절 반점을 분석하기 위해 세 개의 CNN 모델을 사용해 하나로 취합
+ Kawahara and Hamarneh, (2016) : 마찬가지로 각기 다른 해상도의 피부 병변을 분석하기 위해 multi-stream CNN 사용 
+ Gao et al., (2015) : 틈새등현미경으로 찍은 핵성 백내장을 진한하기 위해 pre-trained CNN과 RNN을 연결해 사용

##### 3D Information
3D 정보를 다루는 일도 메디컬계에서 중요하다. 컴퓨터비전이 주로 2D를 다루기 때문에 직접적으로 3D정보를 다루는 쪽으로는 크게 발전하지 못했다.

+ Setio et al., (2016) : 3D 흉부 CT의 결절여부를 분류하기 위해 multi-stream CNN 사용했다. 9개의 서로 다른 patch들이 각각의 stream에서 추출되고, 이후 fully-connected layer에서 취합된다.
+ Nie et al., (2016) : 3D MRI 영상으로 신경교종 환자들의 생존율을 학습하는 일에 3D CNN을 사용했다.

##### RBM, SAE, CSAE..
최근 대부분의 연구들은 CNN을 사용하지만 그렇지 않은 경우도 존재한다.

+ RBM : (van Tulder and de Bruijne, 2016; Zhang et al., 2016)
+ SAE : (Cheng et al., 2016)
+ CSAE : sparse auto-encoder로 unsupervised pre-training을 시키는 점이 CNN과의 다른점(Kallenberg et al., 2016)
+ MIL + DNN: multiple instance learning with deep learning (Xu et al., 2014)

<br>결론적으로 Object Classification은 Exam Classification보다 pre-trained network를 덜 사용한다. contextual 혹은 3D의 정보를 다루기 힘들기 때문이다. 몇몇 저자들은 이 분야에서 DNN으로 혁신적인 해결책을 발견했고 DNN은 가까운 미래에 이 분야에서 더 많이 사용될 것으로 전망된다.

### (2) Detection
#### ① Organ, region and landmark localization

※ Localization : Classification 뿐만 아니라, 해당 Object의 위치를 Bouncing Box 등으로 표시해주는 것

장기, 부위 등의 해부학적 Object Localization은 Segmentation 혹은 치료를 위한 임상계획 업무에 사용되는 중요한 전처리 과정이다. 메디컬 이미지에서의 Localization은 종종 3D 데이터의 파싱을 요구한다. 딥러닝으로 이 문제를 풀기 위해 3D공간을 2D orthogonal plane의 집합으로 처리하는 방법이 제안되었다.

+ Yang et al. (2015) : 원위 대퇴골 표면부를 인식하기 위해 일반적인 CNN에 3개의 independent한 2D MRI 슬라이스 셋들을 사용했다. 부위의 3차원적인 위치는 각각의 2D 슬라이스들이 가장 높은 확률이라고 분류한 위치의 조합으로 선택된다.
+ de Vos et al. (2016) : 3D CT자료를 2D로 파싱해 3D 정육면체로 Bouncing Box를 그려서 해부학적 관심부위(심장, 대동맥궁, 하행 대동맥 등)를 표시했다. 
+ pre-trained CNN이나 RBM도 같은 목적으로 사용된 경우가 있다.

## 5. Discussion

#### Overview

우리가 서베이한 308개의 논문들로부터 딥러닝이 메디컬 이미지 분석의 모든 측면에 사용되고 있다는 것이 명백하게 드러났다. 이런 일들은 정말 빠르게 일어났다. 대다수의 논문들(242편)은 2016년 ~ 2017년 1월 사이에 출간되었고 다양한 딥러닝 아키텍쳐들이 다루어졌다. 


초기 연구들은 **pre-trained CNN**을 feature extractor 정도로 사용했다. 이 pre-trained network들은 쉽게 다운로드 받을 수 있고 곧바로 아무 메디컬 이미지 분석에나 사용될 수 있어서 인기가 많았다. 더욱이 이런 접근법은 handcrafted feature에 기반한 기존 시스템을 간단히 확장시킬 수 있었다. 


하지만 지난 2년 간 **end-to-end trained CNN**이 메디컬 이미지 해석에 더 많이 이용되는 것을 알 수 있었다. 이런 CNN들은 종종 원래 있던 이미지 분석 파이프라인에 추가되거나 기존의 오래된 handcrafted 방식을 대체하였다. 우리 서베이에서도 이 부분이 가장 많이 다루어졌고 그로써 이것이 현재의 표준 방식이라고 말할 수 있다.

---
#### Key aspects of successful deep learning methods

각 업무 및 응용분야 별로 완벽한 딥러닝 방법론과 아키텍쳐를 추출해 낼 수 있을까? CNN은 현재 명백히 대부분의 메디컬 이미지 분석에서 높은 성과를 내고 있고, 우리가 내릴 수 있는 핵심 결론은 좋은 솔루션을 얻는 데에 있어 **가장 중요한 지표는 아키텍쳐가 아니라는 것**이다. 예를 들어 우리는 Kaggle Diabetic Retinopathy Challenge 같은 대회들에서 많은 연구자들이 완전히 똑같은 아키텍쳐, 같은 네트워크를 사용했지만 결과가 많이 다르다는 걸 볼 수 있다. 종종 간과되곤 하는 핵심 양상은, 업무에 대한 **전문적 지식**이 CNN에 layer를 더 추가하는 것보다 효과적일 수 있다는 점이다.


##### 1. Augmentation and Pre-processing

딥러닝 알고리즘을 적용해서 좋은 퍼포먼스를 내는 그룹 및 연구자들은 새로운 ***데이터 전처리 과정***이나 ***확장기술(Augmentation Tech)*** 등을 도입해 차별성을 둔다. CAMELYON16-challenge에서 최고의 퍼포먼스를 낸 팀은 CNN을 바꾸지 않고 Stain Normalization 전처리 과정을 추가해 굉장한 향상(AUC를 0.92에서 0.99로)을 이뤄냈다. 네트워크를 더 robust하게 만들기 위해 Data Augmentation 전략을 적용한 논문들도 있다. (Ronneberger et al., 2015)

물론 Augmentation이나 Pre-processing이 유일한 성능향상 요소는 아니다. 몇몇 연구자들은 업무에 알맞는 특별한 아키텍쳐를 고안해 일반적인 CNN보다 높은 성능을 냈다. Multi-view나 Multi-scale 네트워크가 그 예이다.<br>


##### 2. Input Size & Receptive Field

또 네트워크 디자인에서 종종 저평가되는 요소로서 ***Input size***와 ***Receptive field***가 있다. input size는 문제가 요구하는 resolution이나 context에 맞게 정해져야 한다. 더 많은 context 정보르 얻기 위해 patch size를 키우는 경우도 있지만, 네트워크의 receptive field를 변경하지 않는다면 별로 효과가 없을 수 있다. 만약 해당분야 전문가들도 좋은 퍼포먼스를 이루지 못한다면 네트워크의 input이나 구조를 변경해야 할 수 있다.


##### 3. Hyperparameter Optimization

마지막 요소로는 모델의 ***Hyperparameter Optimization***이 있다. 이건 부차적인 요소이긴 하다. 그리고 최상의 hyperparameter를 찾는 방법도 딱히 없기에 직접 해보는 수밖에 없다. 대부분의 연구자들은 직감에 의지하며 꽤 잘 들어맞기도 한다. 최적화를 위해 Bayesian Method를 사용하기도 했지만 메디컬 이미지 분야에는 적용된 적이 없다.

#### Unique challenges in medical image analysis

##### 1. Training Dataset의 부재

부분적으로만 맞는 말이다. 적어도 10년 전부터 대부분의 서구 병원들은 영상의학(Radiology)에서 PACS 시스템 사용이 일반적이었고 여기서 수백만장의 이미지가 나온다. 다른 몇몇 분야에서도 특정 목적으로 얻어낸 굉장히 많은 양의 이미지 데이터들이 존재한다. PACS같은 시스템은 안과학이나 병리학에서는 널리 사용되고 있지 않지만 점점 사용되는 추세이다. 또한 많은 양의 공공데이터도 존재한다.
+ Esteva et al. (2017) : 18개의 공공데이터 셋으로 \\({10}^{5}\\)개 이상의 training image를 사용
+ Kaggle의 당뇨망막병증 대회에서 이와 비슷한 수의 안구 이미지가 제공됨
+ 몇몇 흉부 엑스레이 연구들도 \\({10}^{4}\\)개 이상의 이미지를 사용


##### 2. Image에 대한 Labeling

그러므로 문제는 데이터셋 자체가 아니라 labeling을 얻는데에 있다. 전통적으로 PACS 시스템은 방사선 전문의들이 그들의 소견을 작성한 text-free 리포트를 저장한다. 이 리포트들을 자동적으로 정확한 주석이나 구조화된 레이블로 만드는 일은 복잡한 텍스트 마이닝 기술을 요구한다. 몇몇 의학계의 structured report의 도입으로 미래에는 데이터 레이블 추출이 더 쉬워질 것이다.

+ Kisilev et al., (2016) : 방사선학에서 BI-RADS 분류체계를 직접적으로 사용해 딥네트워크를 학습
+ Schlegl et al., (2015) : semantic description으로 광학간섭단층촬영 이미지를 분석

우리는 가까운 미래에 네트워크 학습에 free-text나 structured report를 최적으로 사용하는 많은 연구들을 기대한다.

<br>알고리즘 학습에 PACS나 기타 유사 시스템으로부터 free-text report을 활용하는게 어려운 상황에서, 일반적으로 연구자들은 이미지 데이터에 대한 task-specific한 주석을 얻기 위해 각 분야 전문가들을 필요로 한다. 충분히 큰 데이터셋을 labeling하는 것은 굉장히 많은 시간이 들고 이것이 핵심 문제이다. 예를 들어, 보통 3D를 다루는 방사선학에서 segmentation을 하기 위해 딥러닝 시스템을 학습시키기 위해서는 모든 층에 주석을 달아야 하고 많은 시간이 든다. 그러므로 한정된 데이터로 효과적인 학습을 시키는 것이 메디컬 이미지 분석 연구의 중요한 화두이다.

+ Cicek et al., (2016) : 3D 딥러닝 segmentation 시스템을 **sparse**한 2D segmentation만을 사용
+ Yan et al., (2016) : Multiple-instance나 active learning 접근법 사용
+ Rajchl et al., (2016) : Crowd Sourcing을 통한 비전문가의 label을 사용
+ Turkki et al., (2016) : 조직병리학에서 관심 부위를 전문적 경험이 크게 필요하지 않은 특정 면역조직화학상의 반점을 사용해 분석


<br>전문가가 주석처리를 했더라도, 레이블의 노이즈가 상대적으로 적은 컴퓨터비전분야와는 다르게 **label noise**가 중요한 제한요인이 될 수 있다. 예를 들어 폐 CT영상에서 종양탐지를 위한 데이터셋은 LIDC-IDRI가 유명하다. 이 데이터셋에서는 폐의 종양이 네 명의 전문가에 의해 각각 쓰여졌다. 추후 서로 리뷰를 하긴 했지만 의견 일치를 강제하지는 않았다. 결국 만장일치로 종양으로 판단되지 않은 종양들이 그렇지 않은 경우보다 세 배 많았다. 이런 데이터로 딥러닝 시스템을 학습시키는 것은 noise 처리와 표준에 대한 불확실성에 대한 세심한 고려가 요구된다. label의 불확실성에 대한 해결책이 loss function에 있다고 생각할 수도 있겠지만 여전히 도전 과제이다.

<br>메디컬 이미징에서는 종종 classification이나 segmentation을 binary task로 생각한다. (정상 대 비정상, 오브젝트 대 배경) 하지만 높은 확률로 heterogeneous할 수 있기 때문에 너무 큰 단순화일 수 있다. 예를 들어 정상 카데고리는 완전히 정상인 조직으로만 이루어져 있지만, 양성종양 카데고리는 여러 경우로 나눠질 수 있다. 이런 경우 시스템은 가장 일반적인 판단은 내릴 수 있지만 특별한 경우를 놓치고 만다. 해결책은 딥러닝 시스템을 multi-class로 바꾸는 것이다. 하지만 이것 또한 주석에 대한 시간적 문제에 봉착하므로 간단한 일은 아니다. 몇몇 연구자들은 학습과정 자체에 인공지능을 통합시키거나(van Grinsven et al., 2016) hard negative mining을 통해(Wang et al., 2016) 이런 불균형을 처리했다. 하지만 표준 자체에 내재된 노이즈가 있는 경우 이런 전략들은 일반적으로 실패한다.


##### 3. Class Imbalance

메디컬 이미징에서 **비정상 클래스에 대한 이미지 정보를 찾기 힘들** 수 있다. 예를 들어, 전세계 여러 지역에서 유방암 진단 프로그램에 참여해 엄청나게 많은 유방조영상을 마련했다. 하지만 이들 대부분은 정상이었고 의심 증후조차 없었다. 유방조영상에 의심스러운 병소가 있다고 해도 암은 아닌 경우가 많았고, 심지어 대부분의 암세포도 환자를 죽음으로 몰아넣지 않는다. 따라서 딥러닝 연구에 이런 불균형 문제를 해결하는 것도 중요한 연구 분야이다. 일반적으로 업계에서 사용하는 전략은 불충분한 클래스에 대해 data augmentation을 적용하는 것이다. 예를 들어, 새로운 병소를 생성하기 위해 기존 이미지에 scaling이나 rotation을 적용해 사용하는 것이다. 이 전략을 뇌종양 segmentation에 적용한 논문은 다음과 같다. (Pereira et al., 2016)

메디컬 이미지 분석에서 이미지 그 자체로는 중요한 정보를 포함하지 않는 경우가 많다. 내과의사들은 종종 더 나은 진단을 내리기 위해 환자의 진료기록, 나이, 인적사항 등등을 포함시킨다. 몇몇 저자들은 이미 직접적으로 딥러닝에 이 정보들을 적용해 보았다. (Kooi et al., 2017) 하지만 이 저자들은 성능 향상이 기대치에 미치지 못했다고 말했다.

또 하나의 과제 중 하나는 의료 feature의 영향력을 보존하기 위해 이미지상의 몇 천 개의 feature들을 얼마 안 되는 의료 feature들의 갯수에 맞추는 것이다. 내과의사들은 정확한 진단을 위해 해부학적 정보가 필요한 경우가 많다. 하지만 많은 딥러닝 시스템들은 patch classification에만 머무는 상황이다. 즉 patch의 해부학적 위치는 반영되지 않는 것이다. 한 가지 해결책은 이미지 데이터는 통채로 넣고, 학습 시 서로 다른 평가방법을 적용하는 것이다. (Milletari et al., 2016) 이 방법은 또한 메디컬 이미지들의 해부학적 정보는 항상 대략적으로 같은 위치와 크기를 가진다는 상대적으로 static한 특성을 이용했다. 하지만 위에서 언급했듯이 전체 이미지에서 관심영역이 작은 경우에는 효과적이지 못했다. 게다가 이미지를 통채로 네트워크에 넣는 일은 메모리 제한 등의 문제를 발생시킬 수 있다. GPU기술의 발전으로 몇몇 경우를 감당할 수 있겠지만, 기가픽셀 단위의 이미지를 사용하는 경우엔 다른 방법을 강구해야 한다.

#### Outlook

비록 언급된 대부분의 문제들의 대부분이 적절히 해결되지는 못했지만, 높은 수준의 성공을 이룬 몇몇 연구들이 존재한다.

+ Esteva et al. (2017), Gulshan et al. (2016) : 이 두 논문은 특정 작업에서 전문가 수준을 뛰어넘는 성능을 이룩했다. 하지만 전체적으로 보자면 대부분의 작업이 해결되지 않았다는 걸 의미한다. 게다가 두 논문은 컴퓨터비전쪽에 가까운 작은 2D 컬러의 image classification에 집중했고, 매우 잘 라벨링된 데이터셋을 사용했다는 한계가 있다. 메디컬 이미지 작업은 3D gray-scale이나 multi-channel image를 사용하는 경우가 많기 때문이다.
+ Sirinukunwattana et al. (2016) : 핵의 중심부를 바로 예측하는 Non-deep learning method로 classification-based 방식의 성능을 뛰어넘었다.


##### 1. Unsupervised Learning

머신러닝 커뮤니티의 딥러닝쪽 현재 트렌드는 메디컬과도 연관이 되어 있는 unsupervised learning이다. 신경망 르네상스는 2006년 unsupervised 방식의 greedy layerwise pre-training과 함께 시작되었지만, 2012년 ImageNet에서 AlexNet의 성공으로 fully supervised 방식으로 대체되었으며 이 서베이의 대부분도 supervised 접근법을 사용한다. 하지만 unsupervised에 관한 관심도 아직 남아있다.

Unsupervised 방식은 세상에 넘쳐나는 unlabeled data를 바로 학습에 사용할 수 있다는 점에서 매력적이다. 또다른 이유는 이 학습 방식은 사람과 유사하다는 것이다. 우리는 여기에서 특정 라벨 없이 object나 structure를 인식하는 방법을 배울 수 있다. 두 가지 unsupervised 전략이 사용된 논문이 있다. labeled data보다 unlabeled data를 훨씬 더 쉽게 많이 구할 수 있기 때문에 이 논문들은 의미가 있다.

+ Kingma and Welling (2013) : VAE(variational auto-encoder) - Variational Bayesian Graphical Model을 신경망과 조합해 Encoder/Decoder로 사용
+ Goodfellow et al. (2014) : GAN(generative adversarial network) - 두 경쟁적인 CNN 모델을 사용


##### 2. Deep Learning Methods as 'Black Boxes'

마지막으로, 딥러닝은 'Black Box'로 묘사되곤 한다. 특히 의학계에서는 '책임'이 중요하며 중요한 법적 책임을 져야할 수 있다. 시스템은 어떤 방법으로든 결과를 설명할 수 있어야 한다. 이를 위한 몇몇 전략들이 존재하며, 이런 방법들을 메디컬 이미지 분석에 적용하면 의사나 환자들이 딥러닝 어플리케이션을 더 빠르게 받아들일 수 있을 것이다.

+ Zeiler and Fergus (2014) : Deconvolution Network
+ Springenberg et al. (2014) : Guided Back-Propagation
+ Montavon et al. (2017) : Deep Taylor Composition
+ Karpathy and FeiFei (2015) : 네트워크가 무엇을 지각하는지 알기 위해 이미지의 Textual Representation을 예측
+ Kendall and Gal (2017) : Bayesian statistics와 Deep Network를 합쳐 네트워크의 불확실도를 얻음. 이 방법은 내과의들이 신뢰할 수 없는 결과가 나왔을 때 네트워크에 적용해볼 수 있을 것이다.

메디컬 이미징에서 거의 연구되지 않은 Image Reconstruction에 딥러닝을 적용할 수도 있을 것이다. (Wang, 2016) 딥러닝은 그러므로 **메디컬 이미지 분석** 뿐만 아니라 전체적인 **메디컬 이미징**에도 굉장한 영향력을 끼칠 수 있을 것이다.
